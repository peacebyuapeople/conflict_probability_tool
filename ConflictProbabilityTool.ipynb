```python
from together import Together
from concurrent.futures import ThreadPoolExecutor, as_completed
from scipy.stats import t
import numpy as np
import re

TOGETHER_AI = "together.ai"
TOGETHER_MODEL_LLAMA_3_3_70B="meta-llama/Llama-3.3-70B-Instruct-Turbo"

def get_together_model_response(model_name, system_message, user_message):
    client = Together(api_key=TOGETHER_API_KEY)
    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {
                "role": "system",
                "content": system_message
            },
            {
                "role": "user",
                "content": user_message
            }
        ],
        max_tokens=2000,
        temperature=0.7,
        top_p=0.7,
        top_k=50,
        n=1,
        repetition_penalty=1,
        stop=["<|eot_id|>","<|eom_id|>"],
        stream=False
    )
    if len(response.choices) == 1:
        return response.choices[0].message.content
    raise ValueError("The LLM API returned incorrect result.")

def get_model_response(provider_name, model_name, system_message, user_message):
    if provider_name == TOGETHER_AI:
        return get_together_model_response(model_name, system_message, user_message)

    raise ValueError("Unexpected provider name.")

def get_probability_from_response(response):
    print(response)
    match = re.search(r"Отже, ймовірність досягнення миру:\s*(\d+)%", response)
    if match:
        probability = float(match.group(1))  # Витягуємо тільки число
        return probability

    print(response)
    raise ValueError("Can not parse response")

def calc_stats(probs):
    data = np.array(probs)
    confidence_level = 0.95
    n = len(probs)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    standard_error = std / np.sqrt(n)

    # Confidence interval
    t_value = t.ppf((1 + confidence_level) / 2, df=n-1)
    margin_of_error = t_value * standard_error
    confidence_interval = (mean - margin_of_error, mean + margin_of_error)
    return [mean, std, confidence_interval]

def calculate_probability(provider_name, model_name, system_message, user_message, max_workers = 10, desired_results = 10, extra_task_running = 0):
    def get_specific_model_response():
        return get_model_response(provider_name, model_name, system_message, user_message)

    initial_tasks = desired_results + extra_task_running  # Start with N+1 tasks
    results = []

    with ThreadPoolExecutor(max_workers=10) as executor:
        # Submit initial tasks
        tasks = [executor.submit(get_specific_model_response) for _ in range(initial_tasks)]

        while len(results) < desired_results:
            for future in as_completed(tasks):
                tasks.remove(future)  # Remove completed task from the list
                try:
                    response = future.result()
                    prob = get_probability_from_response(response)
                    if prob is not None:
                        results.append(prob)
                        print(f"Success: {prob}")
                    else:
                        print("Failure: No valid probability returned.")
                        extra_task_running -= 1
                except Exception as e:
                    print(f"Exception: {e}")
                    extra_task_running -= 1

                while extra_task_running < 0:
                    print("Submitted a new task for an exception.")
                    new_task = executor.submit(get_specific_model_response)
                    tasks.append(new_task)
                    extra_task_running += 1

                # Stop if we've collected enough results
                if len(results) >= desired_results:
                    break

    average_probability = sum(results) / len(results)
    print(f"Average probability: {average_probability}")
    print(results)
    print(calc_stats(results))

provider_name = TOGETHER_AI
model_name = TOGETHER_MODEL_LLAMA_3_3_70B
system_message = EVALUATE_PEACE_PLAN_UKRAINIAN_SYSTEM_MESSAGE
user_message = PEACE_PLAN_USER_MESSAGE
print(user_message)
calculate_probability(provider_name, model_name, system_message, user_message, max_workers = 10, desired_results=10)
```
